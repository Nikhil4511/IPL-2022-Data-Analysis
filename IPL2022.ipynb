## Number of Matches won by and defending team

```python
import plotly.graph_objects as go
import seaborn as sns

# Fix 1: Correct typo in 'Wickets'
df['won_by'] = df['won_by'].map({'Wickets': 'Chasing', 'Runs': 'Defending'})

# Count values
won_by = df['won_by'].value_counts()
label = won_by.index
counts = won_by.values
colors = ['red', 'lightgreen']

# Fix 2: Use data= not df=
fig = go.Figure(data=[go.Pie(labels=label, values=counts)])

# Optional Fix 3: Fix hoverinfo format
fig.update_layout(title_text="Number of matches won by defending vs chasing")
fig.update_traces(
    hoverinfo='label+percent',
    textinfo='value',
    textfont_size=30,
    marker=dict(colors=colors, line=dict(color='black', width=3)))

# fig.show()
```

## Recommended Decision

```python
df['toss_win_match'] = df['toss_winner'] == df['match_winner']

win_df = df[df['toss_win_match']]

venue_decision_wins = (
    win_df
    .groupby(['venue', 'toss_decision'])
    .size()
    .reset_index(name='wins')
)

pivot_table = venue_decision_wins.pivot(
    index='venue',
    columns='toss_decision',
    values='wins'
).fillna(0)

def recommend_decision(row):
    if row.get('Bat', 0) > row.get('Field', 0):
        return 'Bat First'
    elif row.get('Field', 0) > row.get('Bat', 0):
        return 'Field First'
    else:
        return 'Neutral (No clear advantage)'

pivot_table['Recommended Decision'] = pivot_table.apply(recommend_decision, axis=1)
```
